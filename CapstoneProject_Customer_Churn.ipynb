```python
# 2. Target (Churn) Vs. Marital Status 

count_marital_status = df['Marital_Status'].unique()
count_churn = df['Churn'].unique()
colors = ['dodgerblue', 'gold']  

fig, ax = plt.subplots(figsize=[5, 7], facecolor='#FFFFFF')

# Create a dictionary to store bar heights for each churn category
churn_data = {}

for churn in count_churn:
    churn_data[churn] = []


# Calculate churn counts for each Marital Status
for marital_status in count_marital_status:
    for churn in count_churn:
        churn_count = df[(df['Marital_Status'] == marital_status) & (df['Churn'] == churn)].shape[0]
        churn_data[churn].append(churn_count)

# Plot the stacked bars
bottom = np.zeros(len(count_marital_status))
for churn, color in zip(count_churn, colors):
    ax.bar(count_marital_status, churn_data[churn], bottom=bottom, color=color, label=f"Churn = {churn}", width=0.6)
    bottom += churn_data[churn]

# Add total churn counts at the top of each stack
for i, total in enumerate(bottom):
    ax.text(i, total + 100, int(total), ha='center', size=12, color='black')

# Customize X-axis with MaritalStatus values
for bar in ax.patches:
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # Center of the bar
        bar.get_height() / 2 + bar.get_y(),  # Center of the stack segment
        round(bar.get_height()),  # Height as label
        ha='center', va='center', size=10, color='black'
    )
# Add labels, title, and legend
ax.set_title('Churn Count by Marital Status', fontsize=16)
ax.set_xlabel('Marital Status', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.legend(title='Churn', labels=['Not Churned', 'Churned'], fontsize=12)

# Remove gridlines and background color
ax.grid(False)
ax.set_facecolor('#FFFFFF')

plt.tight_layout()
plt.show()
```
